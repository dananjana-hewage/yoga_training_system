```python
# import libries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import models, layers
import PIL
import cv2
```

Data preprocessing

training image preprocessing

```python
IMAGE_SIZE = 256
BATCH_SIZE = 32
```

```python
# load the dataset 
training_set=tf.keras.utils.image_dataset_from_directory(
     'YogaPoses',
    batch_size=BATCH_SIZE,
    image_size=(IMAGE_SIZE, IMAGE_SIZE),
    shuffle=True,
    seed=123
   
)
```

```python
class_names =training_set.class_names
class_names
```

```python
# Define a data augmentation pipeline 
data_augmentation = tf.keras.Sequential([
    layers.RandomRotation(0.4),
    layers.RandomWidth(0.2),
    layers.RandomHeight(0.2),
   # layers.RandomShear(0.2),
    layers.RandomZoom(0.2),
    layers.RandomFlip("horizontal"),
])
```

```python
# Normalization layer to rescale the images
resize_and_rescale = tf.keras.Sequential([
    layers.Rescaling(1.0 / 255)
])
```

```python
# Function to count augmented images
def count_augmented_images(dataset, augment_function, num_batches=5):
    total_images = 0
    for image_batch, label_batch in dataset.take(num_batches):  # Take a few batches for counting
        augmented_images = augment_function(image_batch)  # Apply the augmentation function
        total_images += augmented_images.shape[0]  # Count how many images in this batch
    return total_images
```

```python
# Apply both augmentation and rescaling to each image in the dataset
training_set = training_set.map(lambda x, y: (resize_and_rescale(data_augmentation(x)), y))
```

```python
# Optionally, use caching and prefetching for improved performance
AUTOTUNE = tf.data.AUTOTUNE
training_set = training_set.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
# Plotting a batch of augmented images to verify
plt.figure(figsize=(10, 10))
for image_batch, label_batch in training_set.take(1):  # Take one batch to display
    for i in range(12):
        ax = plt.subplot(3, 4, i + 1)
        plt.imshow(image_batch[i].numpy())  # Display each augmented image
        plt.title("Class: " + str(label_batch[i].numpy()))
        plt.axis("off")
plt.show()
```
